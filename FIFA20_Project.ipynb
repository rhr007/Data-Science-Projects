```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data_set = pd.read_csv('players_20.csv')

data_set.head()
```

```python
for col in data_set.columns:
  print(col)
```

```python
data_set.shape
```

```python
data_set.describe()
```

```python
data_set['nationality'].value_counts()
```

```python
data_set['nationality'].value_counts()[0:10]
# here it shows 10 records
```

```python
top_5_country = data_set['nationality'].value_counts()[0:5]

plt.bar(top_5_country.keys(), top_5_country, color="g")
plt.grid(True)
plt.show()
```

```python
player_salary = data_set[['short_name', 'wage_eur']]
player_salary.head()
```

```python
player_salary = player_salary.sort_values('wage_eur', ascending = False)
player_salary_top_10 = player_salary.head(10)
player_salary_top_10
```

```python
plt.figure(figsize = (15, 7))
plt.bar(player_salary_top_10['short_name'], player_salary_top_10['wage_eur'], color= ['yellow', 'green', 'pink', 'purple', 'blue', 'black', 'red', 'orange'])
plt.grid(True)
plt.show()
```
