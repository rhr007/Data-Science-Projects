```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('marvel_demo_stats_powers.csv')

dataset.head()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
dataset['Alignment_x'].value_counts()[0:5]
```

```python
dataset[dataset['Alignment_x'] == 'neutral']
```

```python
good = dataset[dataset['Alignment_x'] == 'good']
good.head()
```

```python
good.sort_values(by = 'Speed', ascending = False).head()
```

```python
good_power = good.sort_values(by = 'Power', ascending = False)[['Name', 'Power']]

good_power.head()
```

```python
good_power[good_power['Name'] == "Thor"]
```

```python
total_score = good.sort_values(by = 'Total', ascending = False)[['Name', 'Total']]
total_score.head()
```

```python
plt.bar(total_score['Name'][0:5], total_score['Total'][0:5], color="green")
plt.xticks(rotation = 90)
plt.title("Total Score Bar Chart")
plt.xlabel("Character Name")
plt.ylabel("Total Score")
plt.grid(True)
plt.show()
```

```python
bad = dataset[dataset['Alignment_x'] == 'bad']
bad.head()
```

```python
bad.sort_values(by = 'Power', ascending = False)[['Name', 'Power', 'Total']].head()
```

```python
bad.sort_values(by = 'Intelligence_x', ascending = False)[['Name', 'Intelligence_x', 'Total']].head()

```

```python
bad.columns.values
```

```python
plt.hist(bad['Combat'], color='orange')
plt.title('Distribution of Combat Scores')
plt.xlabel('Combat Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
